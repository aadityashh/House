```python
def Condition (value):
    if value == 'Bad':
        return 0
    elif value == 'Fair':
        return 1
    elif value == 'Okay':
        return 2
    elif value == 'Good':
        return 3
    else:
        return 4 
```

```python
data['Condition of the House'] = data['Condition of the House'].map(Condition)
```

```python
def detect_outliers(dataset, col):
    seg = dataset[col]

    q1 = seg.quantile(0.25)
    q3 = seg.quantile(0.75)

    iqr = q3 - q1
    max_val = max(seg)

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr 

    if max_val > upper_bound:
        print(f'[!]: Outliers present in {col}')


for col in data.columns:
    detect_outliers(data, col)
```

```python
cols_to_treat = ['No of Bedrooms']

def remove_outliers(dataset, col):
    seg = dataset[col]

    q1 = seg.quantile(0.25)
    q3 = seg.quantile(0.75)

    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr 

    def treat(val):
        if val < lower_bound:
            return lower_bound
        elif val > upper_bound:
            return upper_bound
        else:
            return val

    dataset[col] = dataset[col].map(treat)

    return dataset



for col in cols_to_treat:
    data = remove_outliers(data, col)
```

```python
import matplotlib.pyplot as plt

def plot_distribution(dataset, col):
    plt.figure(figsize=(10, 6))
    plt.title(f'Distribution of {col}')
    plt.hist(dataset[col], bins=30, edgecolor='black')
    plt.show()


for col in data.columns:
    plot_distribution(data, col)
```

```python
import numpy as np

def transform (data, col):
    return np.sqrt( np.abs(data[col]) )

for col in data.columns:
    data[col] = transform(data, col)
```

```python
def plot_distribution(dataset, col):
    plt.figure(figsize=(10, 6))
    plt.title(f'Distribution of {col}')
    plt.hist(dataset[col], bins=30, edgecolor='black')
    plt.show()


for col in data.columns:
    plot_distribution(data, col)
```

```python
def get_correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)

    return col_corr


get_correlation(data, 0.8)
```
